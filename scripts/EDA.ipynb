```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
%%time
train = pd.read_csv('../../TPS_2021/input/tabular-playground-series-mar-2021/train.csv')
test = pd.read_csv('../../TPS_2021/input/tabular-playground-series-mar-2021/test.csv')
```

```python
cat = [f'cat{i}' for i in range(19)]
```

```python
train_test_cat = pd.concat([train[cat], test[cat]], axis =0)
```

```python
for i in range(19):
    le = LabelEncoder()
    train_test_cat[f'cat{i}'] = le.fit_transform(train_test_cat[f'cat{i}'].values)

train_test_cat.head()
```

```python
train[cat] = train_test_cat[:train.shape[0]]
```

```python
test[cat] = train_test_cat[train.shape[0]:]
```
